Screenshot 2025-10-02 113306.png

### Stash this file and check out to feature1 branch

```bash
git stash push -m "Stashing develop.txt"
git checkout feature1
```
Screenshot 2025-10-02 113219.png

###  Create new.txt file in feature1 branch, stage and commit this file

```bash
echo "This is new file in feature1 branch" > new.txt
git add new.txt
git commit -m "Added new.txt in feature1 branch"
```
Screenshot 2025-10-02 113102.png
###  Checkout to develop, unstash this file and commit

```bash
git checkout develop
git stash pop
git add develop.txt
git commit -m "Added develop.txt in develop branch"

```
Screenshot 2025-10-02 112948.png

## 3. Tasks to Be Performed:
### 1. Create a Git working directory, with the following branches:
  Develop, F1, f2
  ```bash
  git branch Develop
git branch F1
git branch f2
  ```
  Screenshot 2025-10-02 180809.png
### 2. In the master branch, commit main.txt file
```bash
echo "This is main file" > main.txt
git add main.txt
git commit -m "Added main.txt in main branch"
```
Screenshot 2025-10-02 180709.png

### 3. Put develop.txt in develop branch, f1.txt and f2.txt in f1 and f2 respectively
```bash
git checkout Develop
echo "This is develop branch file" > develop.txt
git add develop.txt
git commit -m "Added develop.txt in Develop branch"

git checkout F1
echo "This is f1 branch file" > f1.txt
git add f1.txt
git commit -m "Added f1.txt in F1 branch"

git checkout f2
echo "This is f2 branch file" > f2.txt
git add f2.txt
git commit -m "Added f2.txt in f2 branch"
```
Screenshot 2025-10-02 180910.png

### 4. Push all these branches to GitHub
```bash
git push -u origin main
git push -u origin Develop
git push -u origin F1
git push -u origin f2
```
Screenshot 2025-10-02 180555.png

### 5. On local delete f2 branch
```bash
git branch -d f2
```
Screenshot 2025-10-02 180348.png

### 6. Delete the same branch on GitHub as well.
```bash
git push origin --delete f2
```
Screenshot 2025-10-02 180301.png

Screenshot 2025-10-02 181646.png

## 4. Tasks to Be Performed:
### 1. Put master.txt on master branch, stage and commit
```bash
echo "This is master file" > master.txt
git add master.txt
git commit -m "Added master.txt in master branch"
```
Screenshot 2025-10-02 183240.png

### 2. Create 3 branches: public 1, public 2 and private
```bash
git branch public1
git branch public2
git branch private
```
Screenshot 2025-10-02 183322.png

### 3. Put public1.txt on public 1 branch, stage and commit
```bash
git checkout public1
echo "This is public1 file" > public1.txt
git add public1.txt
git commit -m "Added public1.txt in public1 branch"
```
Screenshot 2025-10-02 183438.png

### 4. Merge public 1 on master branch
```bash
git checkout master
git merge public1
```
Screenshot 2025-10-02 183529.png

### 5. Merge public 2 on master branch
```bash
git checkout public2
echo "This is public2 file" > public2.txt
git add public2.txt
git commit -m "Added public2.txt in public2 branch"

git checkout master
git merge public2
```
Screenshot 2025-10-02 183755.png

### 6. Edit master.txt on private branch, stage and commit
```bash
git checkout private
echo "Updated master file in private branch" > master.txt
git add master.txt
git commit -m "Updated master.txt in private branch"
```
Screenshot 2025-10-02 183858.png

### 7. Now update branch public 1 and public 2 with new master code in private
```bash
# Update public1
git checkout public1
git merge private

# Update public2
git checkout public2
git merge private
```
Screenshot 2025-10-02 184046.png

### 8. Also update new master code on master
```bash
git checkout master
git merge private

```
Screenshot 2025-10-02 184613.png

### 9. Finally update all the code on the private branch
```bash
git checkout private
git merge master
```
Screenshot 2025-10-02 184317.png

Screenshot 2025-10-02 184518.png

## 5. Tasks to Be Performed:
### 1. Create a Git Flow workflow architecture on Git
```bash
mkdir GitProject6
cd GitProject6
git init -b main
```
Screenshot 2025-10-02 224517.png

### 2. Create all the required branches
```bash
git branch develop
git checkout develop
```
Screenshot 2025-10-02 224430.png

### 3. starting from e feature branch, push the branch to the master, following the architecture
```bash
# Create feature branch from develop
git checkout -b feature/login

# Work on feature
echo "Login feature code" > login.txt
git add login.txt
git commit -m "Added login feature"

# Merge feature into develop
git checkout develop
git merge feature/login

git checkout -b release/v1.0

# (Optional: final fixes)
echo "Release notes v1.0" > release.txt
git add release.txt
git commit -m "Prepared release v1.0"

# Merge release into master
git checkout master
git merge release/v1.0

# Tag the release
git tag -a v1.0 -m "Release version 1.0"

# Merge release back into develop
git checkout develop
git merge release/v1.0

```
Screenshot 2025-10-02 224325.png

### 4. Push an urgent.txt on master using hotfix
```bash
# Checkout master for production hotfix
git checkout master

# Create hotfix branch
git checkout -b hotfix/urgent-fix

# Add urgent file
echo "Urgent fix applied" > urgent.txt
git add urgent.txt
git commit -m "Hotfix: urgent issue fixed"

# Merge hotfix into master
git checkout master
git merge hotfix/urgent-fix
git tag -a v1.0.1 -m "Hotfix urgent issue"

# Merge hotfix into develop too
git checkout develop
git merge hotfix/urgent-fix
```
Screenshot 2025-10-02 224013.png
