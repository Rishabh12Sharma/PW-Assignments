## 2. Describe the role of predefined keywords in Python and provide examples of how they are used in a program.
### --> Predefined Keywords in Python
Keywords in Python are reserved words that have special meanings and are used to define the syntax and structure of the program. These keywords cannot be used as variable names.

🔹 Key Points:
Keywords are predefined and reserved.
They define the structure of Python programs.
Cannot be used as identifiers (variable names, function names, etc.).
🔹 Common Keywords:

if, else, elif, for, while, def, class, return, True, False, None

```bash
# Using keywords in a program
x = 10

if x > 5:
    print("x is greater than 5")
else:
    print("x is smaller or equal to 5")
```

##  3. Compare and contrast mutable and immutable objects in Python with examples.
## --> Mutable vs Immutable Objects
In Python, objects are classified as mutable or immutable based on whether their values can be changed after creation.

🔹 Mutable Objects:

Mutable objects can be modified after they are created.

Examples:
->List
->Dictionary
->Set
```bash
# Mutable example
my_list = [1, 2, 3]
my_list[0] = 10
print(my_list)
```
🔹 Immutable Objects:

Immutable objects cannot be changed once created. Any modification creates a new object.

Examples:
->Integer
->String
->Tuple
```bash
# Immutable example
x = 10
x = 20  # new object created

text = "hello"
# text[0] = 'H'  # This will give an error
```

## 4. Discuss the different types of operators in Python and provide examples of how they are used.
### --> Types of Operators in Python

Operators are used to perform operations on variables and values.

🔹 Types of Operators:
1. Arithmetic Operators
Used for mathematical calculations.
```bash
a = 10
b = 3
print(a + b)   # Addition
print(a ** b)  # Power
```
2. Comparison Operators
Used to compare values.
```bash
print(5 > 3)
print(5 == 3)
```
3. Logical Operators
Used to combine conditions.
```bash
print(True and False)
print(True or False)
```
4. Assignment Operators
Used to assign values.
```bash
x = 5
x += 2
print(x)
```
5. Membership Operators
```bash
print(3 in [1, 2, 3])
```
6. Identity Operators
```bash
a = [1, 2]
b = a
print(a is b)
```

## 5. Explain the concept of type casting in Python with examples.
### --> Type Casting in Python
Type casting refers to converting one data type into another.

🔹 Types:
1. Implicit Casting
Automatically done by Python.
```bash
x = 5
y = 2.5
z = x + y
print(z)
```

## 6. How do conditional statements work in Python? Illustrate with examples.
### --> Conditional Statements in Python
Conditional statements are used to make decisions in a program based on conditions.

🔹 Types:
1. if Statement
```bash
x = 10
if x > 5:
    print("Greater")
```
2. if-else Statement
```bash
x = 3
if x > 5:
    print("Greater")
else:
    print("Smaller")
```
3. if-elif-else Statement
```bash
marks = 75

if marks >= 90:
    print("A")
elif marks >= 70:
    print("B")
else:
    print("C")
```
4. Nested if
```bash
x = 10
if x > 5:
    if x < 20:
        print("Between 5 and 20")
```

## 7. Describe the different types of loops in Python and their use cases with examples.
### --> Loops in Python
Loops are used to execute a block of code repeatedly.

🔹 1. for Loop

Used when the number of iterations is known.
```bash
for i in range(5):
    print(i)
```
🔹 2. while Loop

Used when the condition is based.
```bash
i = 0
while i < 5:
    print(i)
    i += 1
```
🔹 Loop Control Statements
break

Stops the loop.
```bash
for i in range(5):
    if i == 3:
        break
```
continue

Skips current iteration.
```bash
for i in range(5):
    if i == 3:
        continue
    print(i)
```
pass

Does nothing.
```bash
for i in range(5):
    pass
```
