## 1. Research and compare different configuration management tools (Ansible,Puppet, Chef) and write a report on their use cases and advantages?

### ---> 📘 Configuration Management Tools: Ansible, Puppet, Chef
1. Introduction

Configuration management is a crucial part of DevOps that focuses on automating the process of configuring, maintaining, and managing infrastructure and software systems. It ensures that systems remain in a consistent and desired state, reducing manual errors and improving efficiency.

Some of the most widely used configuration management tools are:

-> Ansible

-> Puppet

-> Chef

These tools help automate tasks such as:

Server setup and configuration
Application deployment
Infrastructure provisioning
Continuous integration and delivery

## Ansible
Ansible is an open-source configuration management and automation tool developed by Red Hat. It is known for its simplicity and ease of use.

It uses YAML-based playbooks to define automation tasks and works in an agentless architecture, meaning no software needs to be installed on target machines.


Architecture

-> Control Node: Machine where Ansible is installed

-> Managed Nodes: Target systems

-> Communication via SSH

Key Features


-> Agentless (no installation on client machines)

-> Simple YAML syntax

-> Push-based model

-> Idempotent operations (safe repeated execution)


Advantages

-> Easy to learn and implement

-> Quick setup and deployment

-> Minimal overhead

-> Strong community support


Use Cases

-> Application deployment

-> Cloud provisioning (AWS, Azure)

-> Configuration automation

-> CI/CD pipeline integration


Example Playbook
```bash
- name: Install NGINX
  hosts: web
  become: yes

  tasks:
    - name: Install nginx
      apt:
        name: nginx
        state: present
```

## Puppet
Puppet is a powerful configuration management tool that uses a declarative language to define system configurations.

It follows an agent-based architecture, where nodes regularly check with a central server to maintain their desired state.


Architecture

-> Puppet Master: Central server

-> Puppet Agents: Installed on nodes

-> Communication via secure SSL


Key Features

-> Declarative language

-> Pull-based model

-> Automatic state enforcement

-> Scalable architecture


Advantages

-> Ensures system consistency

-> Highly scalable

-> Strong compliance and reporting

-> Suitable for large infrastructures


Use Cases

-> Enterprise infrastructure management

-> Continuous configuration enforcement

-> Security compliance environments

-> Data center automation


Example Manifest
```bash
package { 'nginx':
  ensure => installed,
}
```

## Chef

Chef is a configuration management tool that uses a Ruby-based DSL (Domain Specific Language).

It is highly flexible and designed for complex infrastructure environments.


Architecture

-> Chef Server: Stores configurations

-> Chef Workstation: Where code is written

-> Chef Nodes: Managed systems


Key Features

-> Uses Ruby DSL

-> Agent-based

-> Supports test-driven development

-> Highly customizable


Advantages

-> Very powerful and flexible

-> Ideal for complex workflows

-> Strong cloud integration

-> Supports Infrastructure as Code


Use Cases

-> Complex enterprise applications

-> Multi-cloud environments

-> Continuous deployment pipelines

-> DevOps automation


Example Recipe
```bash
package 'nginx' do
  action :install
end
```

## 2. Extend the CI pipeline to include continuous delivery using Jenkins or GitLab CI. Implement a blue-green deployment strategy?

--> Continuous Integration (CI) ensures that code changes are automatically built and tested. To improve this, we extend CI into Continuous Delivery (CD), where code is also automatically prepared for deployment.

### What is Continuous Delivery (CD)?
Continuous Delivery ensures:

Code is always in a deployable state
Automated build, test, and deployment stages
Faster and more reliable releases

Pipeline Flow
```bash
Code → Build → Test → Deploy → Production
```

### CI/CD Pipeline using Jenkins
Pipeline Stages
1. Build Stage
Compiles the application
Installs dependencies
2. Test Stage
Runs unit and integration tests
3. Deploy Stage
Deploys application to staging/production

Jenkins Pipeline (Jenkinsfile)
```bash
pipeline {
    agent any

    stages {
        stage('Build') {
            steps {
                echo 'Building application...'
            }
        }

        stage('Test') {
            steps {
                echo 'Running tests...'
            }
        }

        stage('Deploy to Green') {
            steps {
                echo 'Deploying new version to Green environment...'
            }
        }

        stage('Switch Traffic') {
            steps {
                echo 'Switching traffic from Blue to Green...'
            }
        }
    }
}
```

### Blue-Green Deployment Strategy

Concept

Blue-Green deployment uses two identical environments:

-> Blue Environment → Current live version

-> Green Environment → New version

Working Steps

-> Current users are on Blue

-> Deploy new version to Green

-> Test Green environment

-> Switch traffic to Green

-> Blue becomes backup
